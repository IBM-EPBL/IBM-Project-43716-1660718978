# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading Dataset

```python
df = pd.read_csv("dataset_website.csv").set_index('index')
df.head()
```

# Result Data

```python
df.Result.unique() #Unique values in Result col
```

```python
count_Negative = (df['Result'] == -1).sum()
count_Positive = (df['Result'] == 1).sum()
# palette_color  = sns.color_palette("Set2")

plt.pie([count_Negative,count_Positive ], labels=['Negative', 'Positive'],autopct='%.0f%%')
plt.show()
```

# Column data Percentage Distribution

```python
col_list = list(df.columns)
col_list
```

```python
final_data = []
for col in col_list:
    temp_dict = {}
    df_len = len(df)
    negative_count = (df[col] == -1).sum()
    positive_count = (df[col] == 1).sum()
    zeros_count = (df[col] == 0).sum()
    temp_dict['col_name'] = col
    temp_dict['negative_count'] = negative_count
    temp_dict['positive_count'] = positive_count
    temp_dict['zeros_count'] = zeros_count
    temp_dict['negative_count_pct'] = (negative_count/df_len)*100
    temp_dict['positive_count_pct'] = (positive_count/df_len)*100
    temp_dict['zeros_count_pct'] = (zeros_count/df_len)*100
    
    final_data.append(temp_dict)

count_df = pd.DataFrame.from_dict(final_data)
# count_df = count_df.transpose()
# new_header = count_df.iloc[0]
# count_df = count_df[1:]
# count_df.columns = new_header
count_df
```

```python
count_df.plot(x="col_name", y=["negative_count_pct", "positive_count_pct","zeros_count_pct"], kind="bar",figsize=(20, 10))
```
